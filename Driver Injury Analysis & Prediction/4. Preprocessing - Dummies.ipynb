# Create Dummies

```python
import pandas as pd
crash3 = pd.read_csv('crash_v2.csv')
```

# Drop Second impact as they are almost all the same as First Impact

```python
match_count = (crash3['First Impact'] == crash3['Second Impact']).sum()

print(f"Number of rows where 'First Impact' matches 'Second Impact': {match_count}")
```

```python
categorical_columns = crash3.select_dtypes(include=['object']).columns
categorical_columns
```

```python
dummy_collision = pd.get_dummies(crash3['Collision'], prefix='Collision')
dummy_collision
```

```python
dummy_street = pd.get_dummies(crash3['Cross-Street Type'], prefix='Street')
dummy_street
```

```python
dummy_fault = pd.get_dummies(crash3['Driver At Fault'], prefix='Fault')
dummy_fault
```

```python
dummy_parked = pd.get_dummies(crash3['Parked Vehicle'], prefix='Parked')
dummy_parked
```

```python
dummy_turn = pd.get_dummies(crash3['Turn'], prefix='Turn')
dummy_turn
```

```python
dummy_light = pd.get_dummies(crash3['Light Conditions'], prefix='Light')
dummy_light
```

```python
dummy_substance = pd.get_dummies(crash3['Substance'], prefix='Substance')
dummy_substance
```

```python
dummy_distraction = pd.get_dummies(crash3['Distraction'], prefix='Distraction')
dummy_distraction
```

```python
dummy_conditions = pd.get_dummies(crash3['Conditions'], prefix='Conditions')
dummy_conditions
```

```python
dummy_movement = pd.get_dummies(crash3['Movement'], prefix='Movement')
dummy_movement
```

```python
dummy_body = pd.get_dummies(crash3['Body'], prefix='Body')
dummy_body
```

```python
dummy_signage = pd.get_dummies(crash3['Traffic Signs'], prefix='Signage')
dummy_signage
```

```python
dummy_1st_impact = pd.get_dummies(crash3['First Impact'], prefix='1st_Impact')
dummy_1st_impact
```

```python
dummies_concat = pd.concat([dummy_collision, dummy_street, dummy_fault,
                           dummy_parked, dummy_turn, dummy_light, dummy_substance,
                           dummy_distraction, dummy_conditions, dummy_movement, dummy_body,
                           dummy_signage, dummy_1st_impact], axis=1)
```

```python
dummies_concat.head()
```

```python
dummies_concat.to_csv('dummies.csv', index=False)
```
